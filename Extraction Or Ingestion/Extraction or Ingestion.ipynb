### I Will Make The connection To API To get Data about The consumption of all USA States of electricity 

```python
API_KEY = "WRqKD1G6pseUHWrppNQM2OHc7zz3CVdMRcOQ1Qzp"
base_url = "https://api.eia.gov/v2/electricity/retail-sales/data/"
params = {
    "api_key": API_KEY,
    "data[0]": "sales",
    "start": 2005,
    "end": 2023,
    "frequency": "monthly",              
    "length": 5000
}

all_data = []
offset = 0

while True:
    params["offset"] = offset
    r = rq.get(base_url, params=params)
    data = r.json()["response"]["data"]
    if not data:
        break
    all_data.extend(data)
    offset += 5000 

Cons_df = pd.DataFrame(all_data)
print("Total rows:", len(Cons_df))
```

### Trying to get monthy values of GDP from yearly values by interpolation

```python
GDP_df 
```

```python
GDP_df = GDP_df[GDP_df["LineCode"]==3]
```

```python
GDP_df.reset_index()
```

```python
GDP_df["Year-Quarter"] = GDP_df["Year-Quarter"].str.replace(":", "",regex=False)

GDP_df["Quarter"] = pd.PeriodIndex(GDP_df["Year-Quarter"], freq="Q")

GDP_df["date"] = GDP_df["Quarter"].dt.to_timestamp()

GDP_df= (
    GDP_df.set_index("date")
      .groupby("GeoName")["GDP"]
      .apply(lambda s: s.resample("MS").interpolate())
      .reset_index()
)
```
